```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
sns.relplot(data=df,x="total_bill", y="tip")
```

```python
sns.relplot(data=df, x="total_bill", y="tip",hue="smoker")
```

```python
sns.relplot(data=df,x="total_bill", y="tip", hue="smoker", style="smoker")
```

```python
sns.relplot(data=df,x="total_bill",y="tip",hue="smoker",style="time")
```

```python
sns.boxplot(data=df,x="smoker",y="total_bill")
```

```python
sns.countplot(data=df,x="smoker")
```

```python
sns.barplot(data=df,x="total_bill",y="tip")
```

```python
sns.barplot(data=df,y="total_bill",x="smoker", )
```

```python
sns.violinplot(data=df, x="smoker", y="day")
```

```python
sns.displot(data=df, x="smoker", y="sex")
```

```python
sns.relplot(data=df, y="smoker", x="sex", hue="size")
```

```python
sns.boxplot(data=df, y="day", x="sex")
```
